_Заполним пропущенные значения там, где это возможно._

##### 'balcony', 'parks_around3000', 'ponds_around3000'

```python
print('Уникальные значения balcony:', list(data['balcony'].unique()))
print('Уникальные значения parks_around3000:', list(data['parks_around3000'].unique()))
print('Уникальные значения ponds_around3000:', list(data['ponds_around3000'].unique()))
```

Если продавец не указал число балконов, то, скорее всего, в его квартире их нет. Такие пропуски можно заменить на число 0.
Аналогичная история для кол-ва парков и водоёмов parks_around3000, ponds_around3000.

Также заменим тип данных в этом столбце с float64 на int, так как колличество балконов, водоёмов и парков является целой величиной.

```python
data['balcony'] = data['balcony'].fillna(0)
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)

data['balcony'] = data['balcony'].astype('int')
data['parks_around3000'] = data['parks_around3000'].astype('int')
data['ponds_around3000'] = data['ponds_around3000'].astype('int')
```

#####  'is_apartment'
is_apartment — апартаменты  
Данный столбец имеет больше всего пропусков.  
Посмотрим все уникальные значения столбца:  

```python
list(data['is_apartment'].unique())
```

Можно предположить, что жилые помещения, где находятся пропуски не являются аппартаментами, поэтому были пропущены.
Но данный параметр очень важен, ведь аппартаменты по размеру налога и другим характеристикам принципиальные отличия от жилой недвижимости.  

Заменим пропуски на False, но будим иметь ввиду, что анализ по данному столбцу может быть не точен и стоит его избегать.  
Также необходимо произвести преобразование типа данных из object в bool.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'] = data['is_apartment'].astype('bool')
```

##### 'ceiling_height'
'ceiling_height'- высота потолков, это ещё один столбец с пропусками
Пропуски могут возникать т.к. продавец не всегда знает высоту потолков в своей квартире.
Оценим данные из этого столбца:

```python
data['ceiling_height'].describe()
```

Здесь присутствуют аномальные значения(если посмотреть на max), поэтому пропуски лучше всего заполнить медианным значением.

```python
ceiling_height_median = data['ceiling_height'].median()
data['ceiling_height'].fillna(ceiling_height_median, inplace=True)
```

##### 'locality_name'

`locality_name` - название населённого пункта
Названий населённого пункта пропущено довольно маленькое кол-во, а поиск даже по информации об аэропорте, водоёмах и парках может быть не точным, поэтому удобно заполнить значением 'Unknown'.

```python
data['locality_name'] = data['locality_name'].fillna('Unknown')
```

##### 'floors_total'

`floors_total` – всего этажей в доме   
Количество пропусков не большое, всего 86 записей.  
Можно предположить, что это одноэтажные дома. Но на всякий случай посмотрим информацию о недвижимости с пропущенной информацией о кол-ве этажей.

```python
display(data.loc[data['floors_total'].isna()].head(10))
```

```python
data['floors_total'].unique()
```

Предположение о том, что этажность не заполнена т.к. этажей всего один оказалось неверно ('floor' везде разный и почти везде > 1), поэтому заполним пропуски '-1' (отрицательных этажей нет в списке уникальных значений), что в нашей ситуации будет обозначать, что информации нет.
Также заменим тип данных в этом столбце с float64 на int, так как кол-во этажей является целым числом.

```python
#Пока не будем удалять строки с пропусками в данном столбце, 
#когда будем его анализировать просто откинем значения < 0
data['floors_total'] = data['floors_total'].fillna(-1)
data['floors_total'] = data['floors_total'].astype('int')
```

##### 'kitchen_area', 'living_area'

Столбцы `living_area` и `kitchen_area` заполним с помощью медианного значения в зависимости от общей площади недвижимости (общая площадь известна для всей недвижимости).

```python
living_area_median = (data['living_area']/data['total_area']).median()
data['living_area'] = data['living_area'].fillna(data['total_area']*living_area_median)

kitchen_area_median = (data['kitchen_area']/data['total_area']).median()
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*kitchen_area_median)
```

Добавим столбец с типом (категорией) этажа квартиры:
- первый (first)
- последний (last)
- другой (other)
- неизвестно (unknown) не стоит забывать, что для небольшой части недвижимости не было известно кол-во этажей, и если мы не добавим ещё одну категорию, то все строки с неизвестным кол-вом этажей попадут в категорию 'other'

```python
def get_floor_category(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'first'
    elif floor ==  floors_total:
        return 'last'
    elif floors_total ==  -1:
        return 'unknown'
    else:
        return 'other'
    
data['floor_category'] = data.apply(get_floor_category, axis= 1)    
```

#### Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений. 

- Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений.  
- Построим сводную таблицу с количеством объявлений и средней ценой квадратного метра для этих населенных пунктов.  
- Выделим населённые пункты с самой высокой и низкой стоимостью квадратного метра.

Для этого составим сводную таблицу по названиям населенных пунктов 'locality_name' с количеством объявлений 'count' и средней ценой квадратного метра 'mean_price_per_m2' для этих населенных пунктов. Отсортуруем данную таблицу по убыванию по кол-ву объявлений для каждого населённого пунката и выведем первые 10 строк, которые отображают информацию по 10 населённых пунктах с наибольшим числом объявлений. 

```python
mean_price_per_m2_table = data.pivot_table(
    index='locality_name', 
    values='price_per_m2', 
    aggfunc=['count', 'mean']
).rename(
    columns = {'count': 'count', 'mean': 'mean_price_per_m2', 'price_per_m2': ''}
).sort_values(
    by='count', 
    ascending=False).head(10)
mean_price_per_m2_table
```

```python
mean_price_per_m2_table.plot(y='count',  autopct='%1.1f%%', kind='pie', legend=False, 
                             title='Доля объявлений по населенным пунктам')
```

```python
print('Самая высокая стоимость квадратного метра (10 населённых пунктах с наибольшим числом объявлений.):')
mean_price_per_m2_table[mean_price_per_m2_table['mean_price_per_m2'] == mean_price_per_m2_table['mean_price_per_m2'].max()]
```

```python
print('Самая низкая стоимость квадратного метра (10 населённых пунктах с наибольшим числом объявлений.):')
mean_price_per_m2_table[mean_price_per_m2_table['mean_price_per_m2'] == mean_price_per_m2_table['mean_price_per_m2'].min()]
```
